```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df_iris = pd.read_csv('iris.csv')
df_iris
```

# Tugas

# Scatter Plot

```python
plt.scatter(x= df_iris['SepalLengthCm'],label='Sepal', y = df_iris['SepalWidthCm'], color='b')
plt.scatter(x= df_iris['PetalLengthCm'],label='Petal' ,y = df_iris['PetalWidthCm'], color='r')
plt.legend()

plt.show()
```

# Histogram

```python
plt.hist(x=df_iris[['SepalLengthCm', 'SepalWidthCm','PetalLengthCm', 'PetalWidthCm']])
```

# Box Plot

```python
df_iris.boxplot(column=["SepalLengthCm", 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], by=None, ax=None)
```

# Pie Chart

```python
pie_data = df_iris.iloc[:5, :]

fig, axs = plt.subplots(2, 2, constrained_layout=True)

axs[0, 0].pie(pie_data['SepalLengthCm'],labels=pie_data['Species'], autopct='%1.1f%%', shadow=True, startangle=90)

axs[0, 1].pie(pie_data['SepalWidthCm'],labels=pie_data['Species'], autopct='%1.1f%%', shadow=True, startangle=90)

axs[1, 0].pie(pie_data['PetalLengthCm'],labels=pie_data['Species'], autopct='%1.1f%%', shadow=True,startangle=90)


axs[1, 1].pie(pie_data['PetalWidthCm'],labels=pie_data['Species'], autopct='%1.1f%%', shadow=True, startangle=90)

plt.show()
```

```python
plt.figure(0)
plt.title('SepalLengthCm')
plt.pie(pie_data["SepalLengthCm"], labels=pie_data['Species'], autopct='%1.1f%%', shadow=True, startangle=90)

plt.figure(1)
plt.title('SepalWidthCm')
plt.pie(pie_data["SepalWidthCm"], labels=pie_data['Species'], autopct='%1.1f%%', shadow=True, startangle=90)

plt.figure(2)
plt.title('PetalLengthCm')
plt.pie(pie_data["PetalLengthCm"], labels=pie_data['Species'], autopct='%1.1f%%', shadow=True, startangle=90)

plt.figure(3)
plt.title('PetalWidthCm')
plt.pie(pie_data["PetalWidthCm"], labels=pie_data['Species'], autopct='%1.1f%%', shadow=True, startangle=90)

plt.show()
```

# Standar Deviasi

```python
from statistics import stdev
print("The Standard Deviation of SepalLengthCm is % s"
                              %(stdev(df_iris["SepalLengthCm"])))
                               
print("The Standard Deviation of SepalWidthCm is % s"
                              %(stdev(df_iris["SepalWidthCm"])))
                               
print("The Standard Deviation of PetalLengthCm is % s"
                              %(stdev(df_iris["PetalLengthCm"])))
                               
                               
print("The Standard Deviation of PetalWidthCm is % s"
                              %(stdev(df_iris["PetalWidthCm"])))
```
